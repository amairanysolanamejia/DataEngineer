```python
# Acceder a una p√°gina en internet y obtener el titulo y el enlace de sitios interesantes
response = requests.get('https://www.worldbank.org/en/research') #Accedemos a la p√°gina del banco mundial
soup = bs4.BeautifulSoup(response.text, 'html.parser')

research = soup.select('.lp__card_content a')                    #Las investigaciones se encuentran en este apartado del 
                                                                 #html

for item in research:                                            #Muestra en pantalla el t√≠tulo de la investigaci√≥n
    print('Titulo: {}'.format(item.text))                        #y el enlace de cada una
    print('Enlace: {}'.format(item['href']))
```
